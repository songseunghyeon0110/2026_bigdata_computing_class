# 1. 넘파일 다차원 배열(ndarray) 객체 생성

- 리스트을 이용하여 다차원 배열 생성
- np.array(리스트)

# 6. 다차원 배열의 지정된 축 방향으로 원소들을 뒤집기
  - np.flip(배열객체, axis=)

```python
# (4, 5) 구조의 이미지 데이터를 생성 -> 상하, 좌우 반전시켜 새로운 이미지 생성
import numpy as np

# 데이터 생성
img = np.arange(20).reshape((4, -1))
print(img)

# (1) 좌우 반전(axis = 1)
h_flip = np.flip(img, axis = 1)
print(h_flip)

# (2) 상하 반전 (axis = 0)
v_flip = np.flip(img, axis = 0)
print(v_flip)

# (3) 전체 반전 (상하 좌우 반전)
all_flip = np.flip(img)
print(all_flip)
```

# 8. 넘파이 다차원 배열의 인덱싱을 이용한 슬라이싱
- 다차원 배열에서 작은 배열을 추출
- 인덱싱을 이용하여 슬라이싱 범위 지정
- 두 축을 이용해서 특정한 행과 열만 별도로 적용하여 추출함
  - axis 0 ,  axis 1
  - arr[] ; axis 0 사용
  - arr[:, :] ;  두 개의 축을 동시에 사용할 때는 쉼표로 구분
  - arr[: :step] ; axis 0을 이용하여 간격을 두고 추출
  - arr[::step, ::step]; 두 개의 축을 이용하여 간격을 두고 추출


```python
# ex : 3차원 배열 슬라이싱
import numpy as np

# 데이터 생성 : 60개 원소로 구성괸 (3, 4) 형태의 3차원 배열 생성
tensor = np.arange(60).reshape((3, 4, 5))
print(tensor)
print("-----------------------------------------------------------")

# (1) 특정 층 전체 가져오기 (첫 번째 층)
print(tensor[0, :, :]) # tensor[0] 이랑 동일 -> shape : (4, 5), size = 20
print("-----------------------------------------------------------")

# (2) 모든 층에서 특정 '행'만 가져오기 (모든 층의 첫 번째 행)
print(tensor[:, 0, :]) # shape : (3,5), size = 15
print("-----------------------------------------------------------")

# (3) 모든 층에서 왼쪽 벽면(첫 번째 열)
print(tensor[:, :,0]) # shape : (3,4), size = 12
print("-----------------------------------------------------------")

# (4) 정가운데 부분만 즉 큐브형태로 도려내기
print(tensor[1:3, 1:3, 1:4])
print("-----------------------------------------------------------")
```

# 9.  넘파이 다차원 배열의 슬라이싱을 이용한 논리인덱싱
- 특정한 조건을 만족하는 배열의 원소값만을 추출하여 1차원 배열 객체로 반환
- 브로드캐스팅하여 병렬 연산 처리


  1. 주어진 배열을 이용하여 조건식을 만족하는 동일한 형상을 가지는 불리언 배열을 생성

  2.   주어진 배열에 불리언 배열을 인덱스로 지정
- cf. np.where(조건식, a, b)
  - 주어진 배열의 구조는 유지하면서 조건식에 맞는 원소를 추출 하는 기법
  - if ~ else 처럼 동작
  - a는 조건식이 true인 경우 처리
  - b는 flase 인 경우 처리






```python
import numpy as np

# 2차원 배열로 데이터 생성
arr = np.arange(100).reshape(5, -1) #(5, 20)

print(arr)

# ex1 : 주어진 배열에서 짝수들만 추출하여 1차원 배열로 생성
# (1) 불리언 배열
arr_bool = arr % 2 == 0 # (5, 20)

print(arr_bool)

# (2) 불리언 인덱싱
arr_even = arr[arr_bool]
print(arr_even) # 1차원 배열

# ex2 : 주어진 배열에서 20보다 큰 요소들만 추출하여 1차원 배열로 생성
over20arr = arr[arr > 20]
print(over20arr)

# ex3 : 주어진 배열에서 20보다 큰 요소들만 추출
arr_filter = np.where(arr > 20, arr, 0)
print(arr_filter)
```

      - NumPy에서 리덕션이 강력한 이유는 **논리값(True/False)**을 연산하는 순간 **숫자(1/0)**로 자동 변환.

      - np.sum(condition) : 조건이 참(True)인 개수를 구함

      - np.mean(condition) : 조건이 참(True)인 비율을 구함 (예: 0.7이면 70%가 참)
      - 이 리덕션 기법들을 활용해서 실제 데이터셋에서 "이상치(Outlier)"를 찾기

# 15. 넘파일 배열의 결합
- np.concatenate(튜플)
- np.vstack(튜플 또는 리스트)
- np.hstack(튜플 또는 리스트)
- np.column_stack(리스트)


```python
#------------------------------------------
# ex1 1차원 배열간 결합
#------------------------------------------
import numpy as np

# 데이터 생성
a = np.array([1, 2, 3]) # (3,)
b = np.array([4, 5, 6]) # (3,)

# 세로로 쌓기 (수직 방향) -> 행 추가 해서 2차원 배열
res_stack = np.vstack((a, b))
print(res_stack) # (2, 3)

# 가로로 붙이기 (수평 방향) -> 열 추가해서 1차원 배열
res_hstack = np.hstack((a, b))
print(res_hstack) # (6,)
res_conc = np.concatenate((a, b), axis = 0)
print(res_conc) # (6,)

# 주의 : column_stack() -> 2차원 배열
res_col = np.column_stack((a, b))
print(res_col) # (3, 2)

print("-------------------------------------")
#------------------------------------------
# ex2 형태가 다른 배열간 결합
#------------------------------------------
arr_1 = np.arange(1, 4) # (1, 3)
arr_2 = np.arange(4, 10).reshape(2,3) # (2, 3)
print(arr_1)
print()
print(arr_2)
print("방법 1")

# 방법 1
# (1) arr_1 -> (1, 3) : 열의 개수를 일치
arr_1_2d = arr_1.reshape(-1, 3)
print(arr_1_2d)
print()
combine_2d = np.concatenate((arr_1_2d, arr_2), axis = 0)
print(combine_2d)
print("방법 2")

# 방법2
v_combine = np.vstack((arr_1, arr_2))
print(v_combine)
```
